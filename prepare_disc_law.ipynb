```python
import json
from pathlib import Path
from collections import Counter

root = Path(r"D:\Projects\Legal-Rag")
raw_dir = root / "DISC-Law-SFT"
out_dir = root / "lf_data" / "disc_law_processed"
out_dir.mkdir(parents=True, exist_ok=True)

files = {
    "pair_full": raw_dir / "DISC-Law-SFT-Pair.jsonl",
    "pair_qa": raw_dir / "DISC-Law-SFT-Pair-QA-released.jsonl",
    "triplet_full": raw_dir / "DISC-Law-SFT-Triplet-released.jsonl",
    "triplet_qa": raw_dir / "DISC-Law-SFT-Triplet-QA-released.jsonl",
}

stats = {}

def read_jsonl(path):
    with path.open("r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if line:
                yield json.loads(line)

# 1. 先统计每个文件的数量和字段
for name, path in files.items():
    count = 0
    key_counter = Counter()
    first_item = None
    
    for item in read_jsonl(path):
        count += 1
        key_counter.update(item.keys())
        if first_item is None:
            first_item = item
    
    stats[name] = {
        "path": str(path),
        "count": count,
        "keys": sorted(key_counter.keys()),
        "first_item": first_item,
    }

for name, info in stats.items():
    print("=" * 100)
    print(name)
    print("path:", info["path"])
    print("count:", info["count"])
    print("keys:", info["keys"])
    print("first item preview:")
    print(json.dumps(info["first_item"], ensure_ascii=False, indent=2)[:1200])

print("=" * 100)
print("TOTAL RAW ROWS:", sum(info["count"] for info in stats.values()))
```

```python
import json
import random
from pathlib import Path
from collections import Counter

random.seed(42)

root = Path(r"D:\Projects\Legal-Rag")
raw_dir = root / "DISC-Law-SFT"
out_dir = root / "lf_data" / "disc_law_processed"
out_dir.mkdir(parents=True, exist_ok=True)

files = {
    "pair_full": raw_dir / "DISC-Law-SFT-Pair.jsonl",
    "pair_qa": raw_dir / "DISC-Law-SFT-Pair-QA-released.jsonl",
    "triplet_full": raw_dir / "DISC-Law-SFT-Triplet-released.jsonl",
    "triplet_qa": raw_dir / "DISC-Law-SFT-Triplet-QA-released.jsonl",
}

sample_limits = {
    "pair_full": 20_000,
    "pair_qa": 40_000,
    "triplet_full": None,  # 全部保留
    "triplet_qa": None,    # 全部保留
}

def read_jsonl(path):
    with path.open("r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if line:
                yield json.loads(line)

def count_file(path):
    return sum(1 for _ in read_jsonl(path))

for name, path in files.items():
    print(name, count_file(path), list(next(read_jsonl(path)).keys()))
```

```python
def normalize_pair(item):
    question = item.get("input", "").strip()
    answer = item.get("output", "").strip()

    if not question or not answer:
        return None

    new_input = (
        "请作为中文法律助手，回答下面的法律问题。回答应当清晰、准确、简洁。\n\n"
        f"【用户问题】\n{question}"
    )

    return {
        "input": new_input,
        "output": answer,
    }

def to_text(value):
    """
    把不同类型的字段统一转成字符串：
    - str: 直接 strip
    - list: 每一项转字符串后用换行拼接
    - dict: 转成 JSON 字符串
    - None: 空字符串
    """
    if value is None:
        return ""

    if isinstance(value, str):
        return value.strip()

    if isinstance(value, list):
        parts = []
        for x in value:
            if isinstance(x, str):
                parts.append(x.strip())
            elif isinstance(x, dict):
                parts.append(json.dumps(x, ensure_ascii=False))
            else:
                parts.append(str(x))
        return "\n".join([p for p in parts if p])

    if isinstance(value, dict):
        return json.dumps(value, ensure_ascii=False)

    return str(value).strip()


def normalize_triplet(item):
    question = to_text(item.get("input", ""))
    answer = to_text(item.get("output", ""))
    reference = to_text(item.get("reference", ""))

    if not question or not answer:
        return None

    if reference:
        new_input = (
            "请根据以下法律依据回答问题。回答应当先给出结论，再说明法律依据和理由。"
            "如果法律依据不足，请明确说明不能确定，不要编造。\n\n"
            f"【法律依据】\n{reference}\n\n"
            f"【用户问题】\n{question}"
        )
    else:
        new_input = (
            "请作为中文法律助手，回答下面的法律问题。回答应当清晰、准确、简洁。\n\n"
            f"【用户问题】\n{question}"
        )

    return {
        "input": new_input,
        "output": answer,
    }
```

```python
output_path = out_dir / "disc_law_mixed_rag_100k.jsonl"

seen = set()
source_counts = Counter()
written = 0

with output_path.open("w", encoding="utf-8") as f_out:
    for name, path in files.items():
        items = list(read_jsonl(path))

        limit = sample_limits[name]
        if limit is not None and len(items) > limit:
            items = random.sample(items, limit)

        for item in items:
            if name.startswith("pair"):
                new_item = normalize_pair(item)
            else:
                new_item = normalize_triplet(item)

            if new_item is None:
                continue

            key = (new_item["input"], new_item["output"])
            if key in seen:
                continue

            seen.add(key)
            f_out.write(json.dumps(new_item, ensure_ascii=False) + "\n")

            source_counts[name] += 1
            written += 1

print("saved to:", output_path)
print("written total:", written)
print("source counts:", dict(source_counts))
```
